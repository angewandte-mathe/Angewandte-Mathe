```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sympy import sign,Function, Derivative
from numpy import abs

fig, ax = plt.subplots(figsize=(12, 5))

m = 0.2
D = 2
z0 = [1,0]

t = np.linspace(0,15,200)

def dF(z,t):
    return[ z[1], -(k/m)* abs(z[1])** n * sign(z[1]) - (D/m)*z[0] ] 

n = 0
k = 1
z = odeint(func=dF, y0=z0, t=t)
n = 0.5
k = 0.1
w = odeint(func=dF, y0=z0, t=t)
n = 1
k = 0.2
e = odeint(func=dF, y0=z0, t=t)
n = 2
k = 0.5
r = odeint(func=dF, y0=z0, t=t)

plt.plot(t, z[:,0],"r", label="n=0, k=1")
plt.plot(t, w[:,0],"g", label="n=0.5, k=0.1")
plt.plot(t, e[:,0],"b", label="n=1, k=0.2")
plt.plot(t, r[:,0],"y", label="n=2, k=0.5")

plt.legend()
plt.grid(True)
plt.show()
```
